```python
marks=np.array([[78,85,90],[45,60,55],[88,92,95],[30,40,35],[65,70,75]])
print(marks)
```

```python
print(marks.shape)
print(marks.size)
print(marks.dtype)
```

```python
print(marks[1,0:3])
print(marks[:,2:])
print(marks[0,1])
```

Analysis


```python
print(np.mean(marks,axis=0))
print(np.mean(marks,axis=1))
print(np.mean(marks[marks>=80]))
print(marks+5)
```

Conditional Logic


```python
print(np.where(marks>=50,"1","0"))
print(np.where(marks<40,"0",marks))
```

Performance Scoring


```python
av=np.mean(marks,axis=1)
print(av)
print(np.where(av>=85,"A",np.where(av>=70,"B",np.where(av>=50,"C","NO grade"))))
```

```python
avgperhos=np.mean(arr1,axis=1)
max_perhos=np.max(arr1,axis=1)
highest_perhos=np.max(avgperhos)
print(avgperhos)
print(max_perhos)
print(highest_perhos)
```

```python
total_per_day=np.sum(arr1,axis=0)
print(total_per_day)
highest_day=np.argmax(total_per_day)
print(highest_day)
```

```python
penalized_usage=np.where(arr1>500,arr1*1.20,arr1)
print(penalized_usage)
```
