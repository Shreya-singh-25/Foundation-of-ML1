```python
import pandas as pd
import numpy as np
```

GroupBy in Pandas

```python
data={
    "Name":["Aman","Bob","Charlie","David","Eve",None,"Sameer"],
    "Age":[25,30,None,40,45,24,28],
    "Marks":[85,90,78,None,92,88,80],
    "Subject":["Math","Science","History","Math",None,"Social Science","Science"]
}
df=pd.DataFrame(data)
df
```

```python
#basic groupby
df.groupby("Subject")['Marks'].mean()
```

```python
df.groupby("Subject")["Marks"].count()
```

```python
#Multiple Aggregations
df.groupby("Subject")["Marks"].agg(["mean","max","min"])
```

```python
#adding count here to see how many students are there in each subject along with mean, max and min marks
df.groupby('Subject')['Marks'].agg(['mean', 'max', 'min', 'count']) 
```

```python
#Group by multiple columns
df.groupby(['Subject', 'Age'])['Marks'].mean()
```
