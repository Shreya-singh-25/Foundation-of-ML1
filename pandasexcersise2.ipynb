```python
import numpy as np
import pandas as pd
```

TechCorp HR Data

```python
hr_data = {"Name":["John","Alice","Bob",None,"Eve","Charlie","David",None,"Aman","Anand"],
           "Department":["HR","Sales","IT","HR","Finance","Sales","Finance","HR","IT","Sales"],
           "Role":["Manager","Executive","Analyst","Executive","Developer","Manager","Analyst","Developer","Executive","Manager"],
           "Experience":[5,3,2,4,8,6,1,2,3,7],
           "Salary":[70000,50000,40000,60000,90000,80000,35000,45000,55000,75000],
           "Score":[85,90,None,88,92,None,75,82,89,91]
           }
df = pd.DataFrame(hr_data)
df
```

excersise ques part1

Basics Averages

```python
df.groupby("Department")["Salary"].mean()
```

```python
df.groupby("Role")["Score"].mean()
```

Counting and Sizing

```python

df.groupby("Department")['Score'].count()
```

```python
df.groupby("Department").size()
```

excercise ques part2

multiple aggregations

```python
df.groupby("Department")['Salary'].agg(['mean','max','min','count'])
```

```python
df.groupby('Role')['Score'].agg(['mean','max','min','count'])
```

```python
#average salary grouped by both department and role
df.groupby(['Department','Role'])['Salary'].mean()
```

```python
#average performance score grouped by both department and experience level
df.groupby(['Department','Experience'])['Score'].mean()
```
