Filtering in pandas

```python
df[df["Marks"]>85]#gives the rows where marks are greater than 85
```

```python
df[(df["Marks"]>90) & (df["Age"]>22)] #gives the rows where marks are greater than 90 and age is greater than 22
```

```python
df.sort_values("Marks") #it will sort the dataframe based on marks in ascending order
```

```python
df.sort_values("Marks",ascending=False) #sort the dataframe based on marks in descending order
```

```python
df["Age"].value_counts() #gives the count of unique values in age column
```

Handling missing values


```python
#updating the dataset with null values

df=pd.DataFrame({"Name":["Saurav","Karan","Aman","Subham","abhishek","rahul"],
                 "Age":[22,23,21,24,25,26],
                 "Marks":[85,90,88,92,87,95]
                 })
df
```

```python
df.isnull() #gives true for null values and false for non null values
```

```python
df.isnull().sum() #gives the count of null values in each column
```

```python
df.dropna() #it will drop the rows with null values
```
